```python
import numpy as np
from scipy.spatial import distance

pointA = np.array([3, 7, 2])
pointB = np.array([8, 4, 5])

# Euclidean Distance
ed = distance.euclidean(pointA, pointB)
print("Euclidean Distance:", ed)

# Similarity (inverse of distance)
s_ed = 1 / (1 + ed)
print("Euclidean Similarity:", s_ed)

# Manhattan (City Block) Distance
manhattan = distance.cityblock(pointA, pointB)
print("Manhattan Distance:", manhattan)

# Similarity (inverse of distance)
s_man = 1 / (1 + manhattan)
print("Manhattan Similarity:", s_man)

# Minkowski Distance (p=3)
minkowski = distance.minkowski(pointA, pointB, 3)
print("Minkowski Distance (p=3):", minkowski)

# Similarity (inverse of distance)
s_min = 1 / (1 + minkowski)
print("Minkowski Similarity (p=3):", s_min)

# Minkowski Distance (p=2)
minkowski2 = distance.minkowski(pointA, pointB, 2)
print("Minkowski Distance (p=2):", minkowski2)

# Similarity (inverse of distance)
s_min2 = 1 / (1 + minkowski2)
print("Minkowski Similarity (p=2):", s_min2)
```

```python
import pandas as pd
from scipy.stats import spearmanr

df = pd.DataFrame({
    'X': [2, 4, 6, 8, 10],
    'Y': [3, 9, 12, 18, 21]
})

corr_value, p_value = spearmanr(df['X'], df['Y'])

print("Spearman correlation coefficient:", corr_value)
print("p-value:", p_value)
```

```python
def hamming_distance(str1, str2):

    # Ensure strings are of equal length
    if len(str1) != len(str2):
        raise ValueError("Strings must be of equal length")

    # Count differing positions
    return sum(ch1 != ch2 for ch1, ch2 in zip(str1, str2))

# Example usage
s1 = "parinitha"
s2 = "parinitah"

distance = hamming_distance(s1, s2)
print("Hamming distance between", s1, "and", s2, "is", distance)
```

```python
# Cosine Similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

s1 = "machine learning is powerful"
s2 = "learning machines can solve problems"

vectorizer = CountVectorizer().fit([s1, s2])
vectors = vectorizer.transform([s1, s2])

cos_sim = cosine_similarity(vectors[0], vectors[1])[0][0]

print("Cosine similarity:", cos_sim)
```

```python
# Sequence-based measures
def lcs_length(X, Y):
    m, n = len(X), len(Y)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m):
        for j in range(n):
            if X[i] == Y[j]:
                dp[i + 1][j + 1] = dp[i][j] + 1
            else:
                dp[i + 1][j + 1] = max(dp[i][j + 1], dp[i + 1][j])

    return dp[m][n]

seq1 = "datascience"
seq2 = "database"

length = lcs_length(seq1, seq2)
print("Length of LCS:", length)
```
